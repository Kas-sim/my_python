```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('updated_yt_data.csv')
print(df)
```

# Red, Blue, Green, Orange, Purple

```python
new_df = df.drop('Video_ID', axis=1)
print(new_df)
```

```python
color_df = pd.get_dummies( new_df['Thumbnail_Color'], prefix='Color',  dtype=int)
print(color_df)
```

```python
upd_df = pd.concat( [ new_df.drop('Thumbnail_Color', axis=1), color_df ], axis=1 )

print(upd_df)
```

```python
upd_df.describe()
```

```python

X = upd_df.drop('Views', axis=1)

# print(X)
Y = upd_df['Views']
# print(Y)
# print(X.shape)
X.describe()
```

```python
model = LinearRegression()
model.fit(X, Y)
```

```python
X_test = [ [ 69, 13, 6969, 0, 0, 1, 0, 0 ] ]
X_test2 = [ [ 24, 17, 60969, 0, 0, 0, 0, 1 ] ]
```

```python
Y_pred = model.predict(X_test2)
print(f"\nyour video will get -> {Y_pred.round()} views!\n")
```
